# Billionaires listed in Forbe

**Dataset Description:**

The Forbes 2640 Billionaire dataset provides an in-depth look into the financial and personal attributes of individuals who have amassed significant wealth and have been recognized by Forbes for their achievements. Each entry in the dataset represents a billionaire, detailing their rank, name, net worth, age, country of origin, industry they are associated with, their primary source of wealth, educational background, marital status, and various other characteristics. The dataset also dives into more nuanced attributes such as the individual's philanthropy score, which might offer insights into their charitable contributions, and the self-made score, giving a perspective on their journey to their current financial standing.

The dataset used in this analysis, titled "Forbes 2640 Billionaire dataset," has been sourced from Kaggle. It can be accessed from the following link: https://www.kaggle.com/datasets/sujalluhar/billionaires-listed-in-forbes

**Research Question:**

Given the diverse range of attributes provided by this dataset, a compelling research question would be:

"How do net worth, age, and country of origin relate to education levels among billionaires?"

This research question aims to uncover patterns or relationships between a billionaire's wealth, age, country of origin, and their educational achievements. It will be particularly interesting to investigate if certain education levels or institutions consistently correlate with higher net worths, or if billionaires from specific countries tend to have a common educational background. Exploring these relationships can provide insights into the potential role of education in amassing significant wealth and how it varies across different regions and age groups.



## Loading important libraries

```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

## Reading the kaggle csv file on billionaire dataset

```python
df = pd.read_csv("/content/forbes_2640_billionaires.csv")
df.drop("age", axis=1)
df
```

### Describe the dataset's dimensions, attributes, and data types

```python
# Dimensions of the dataset
print(f"The dataset has {df.shape[0]} rows and {df.shape[1]} columns.\n")

# Attributes of the dataset
print("The attributes (columns) of the dataset are:", df.columns.tolist(), "\n")

# Data types of each attribute
print("Data types of the attributes:\n", df.dtypes)
```

checking the missing values now


```python
# Check for missing values
missing_values = df.isnull().sum()

print("Number of missing values for each column:\n")
print(missing_values)
```

Filling missing values with the mean

```python
# Loop through each column and fill missing values with the mean if the column contains numeric data
for column in df.columns:
    if df[column].dtype in ['int64', 'float64']:
        df[column].fillna(df[column].mean(), inplace=True)

# Using a for loop convert all floats to ints
for column in df.columns:
    if df[column].dtype == 'float64':
        df[column] = df[column].astype(int)
```

#### Age vs Net Worth

```python
# Scatter plot of Age vs. Net Worth using Matplotlib
plt.figure(figsize=(12, 8))
plt.scatter(df['age'], df['net_worth'], alpha=0.6, edgecolor='black')
plt.title('Age vs. Net Worth using Matplotlib')
plt.xlabel('Age')
plt.ylabel('Net Worth (in billions)')
plt.grid(True)
plt.show()
```

We've already visualized the scatter plot of Age vs. Net Worth. To infer, we can calculate the correlation coefficient to understand the strength and direction of the linear relationship between the two

A value close to 1 implies a strong positive correlation: as one variable increases, the other also tends to increase. A value close to -1 implies a strong negative correlation: as one variable increases, the other tends to decrease. A value close to 0 implies weak correlation: changes in one variable do not predict changes in another variable.

```python
correlation = df['age'].corr(df['net_worth'])
print(f"Correlation between Age and Net Worth: {correlation:.2f}")
```

A correlation coefficient of 0.07 indicates a very weak positive linear relationship between Age and Net Worth.

This means that as age increases, net worth also slightly increases, but the relationship is weak. In practical terms, age alone doesn't significantly determine the net worth of a billionaire in this dataset.

#### Analysis of Net Worth vs. Country

We can group the data by country and compute the average net worth for each country. This will give us an insight into which countries have the highest average billionaire net worth.



```python
average_net_worth_country = df.groupby('country')['net_worth'].mean().sort_values(ascending=False)

# Visualize using Seaborn
plt.figure(figsize=(12, 8))
sns.barplot(x=average_net_worth_country.head(10).index, y=average_net_worth_country.head(10).values)
plt.title('Top 10 Countries by Average Net Worth using Seaborn')
plt.ylabel('Average Net Worth (in billions)')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.show()
```

#### Analysis of Net Worth vs. Education
We can visualize the distribution of net worth across different education levels.

```python
# Using Matplotlib
df.groupby('Education')['net_worth'].mean().sort_values().plot(kind='barh', figsize=(12, 8))
plt.title('Average Net Worth by Education Level using Matplotlib')
plt.xlabel('Average Net Worth (in billions)')
plt.ylabel('Education Level')
plt.show()

# Using Seaborn
plt.figure(figsize=(12, 8))
sns.barplot(y='Education', x='net_worth', data=df, estimator=sum)
plt.title('Average Net Worth by Education Level using Seaborn')
plt.xlabel('Average Net Worth (in billions)')
plt.ylabel('Education Level')
plt.show()
```

let's create a pie chart to analyze bachelors vs masters vs drop out vs others

**Bachelor's Degree:** Those with a bachelor's degree but not a master's.

**Master's Degree:** Those with a master's degree (this group can also include those with a bachelor's degree since typically one gets a master's after a bachelor's).

**Drop Out:** Those who dropped out.

**Others:** Those who don't fit into the above categories.

```python
# Create a new column 'Education_Category' to represent the highest educational qualification
conditions = [
    (df['Master'] == 1),
    (df['Bachelor'] == 1) & (df['Master'] == 0),
    (df['Drop Out'] == 1),
]
choices = ['Masters', 'Bachelors', 'Drop Out']

df['Education_Category'] = np.select(conditions, choices, default='Others')

# Grouping data by 'Education_Category' column
education_counts = df['Education_Category'].value_counts()
education_net_worth = df.groupby('Education_Category')['net_worth'].sum()

# Pie chart for count of billionaires based on highest educational qualification
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
labels = education_counts.index.tolist()
plt.pie(education_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#FF9999', '#66B2FF', '#99FF99', '#FFCC99'])
plt.title('Distribution of Billionaires based on Education')

# Pie chart for net worth of billionaires based on highest educational qualification
plt.subplot(1, 2, 2)
plt.pie(education_net_worth, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#FF9999', '#66B2FF', '#99FF99', '#FFCC99'])
plt.title('Net Worth Distribution based on Education')
plt.tight_layout()
plt.show()
```

**Distribution of Billionaires Based on Education:**

**Dropouts:** Only 3% of the billionaires on the Forbes list are dropouts. This indicates that while formal education might provide certain advantages, there are unique paths to success that don't necessarily require a traditional educational background.

**Masters:** Approximately 19.8% of the billionaires have a master's degree. This indicates a significant number of billionaires pursued further education beyond a bachelor's degree, possibly indicating the value they place on advanced knowledge and skills.

**Bachelors:** The largest group with a formal degree, 24.8% of the billionaires have a bachelor's degree. This suggests that a significant portion of billionaires have at least an undergraduate education.

**Others:** A whopping 52.4% fall into the 'Others' category. This could be those who might have other forms of education, certifications, or no formal education at all but didn't drop out from any institution.

**Net Worth Distribution Based on Education:**

**Others:** This group holds 32.4% of the total net worth of the listed billionaires. Given that they form the majority in terms of count, their combined net worth is substantial but not the highest, indicating a diverse range of net worth values within this group.

**Dropouts:** Interestingly, even though dropouts form only 3% of the billionaires, they hold a massive 39.0% of the total net worth. This suggests that while they are fewer in number, on average, they have a higher net worth than their counterparts. Notable dropouts like Bill Gates and Mark Zuckerberg might be skewing this percentage.

**Masters:** Holding 22.6% of the total net worth, those with master's degrees have a significant financial footprint. This group has a higher average net worth than those with only a bachelor's degree but less than dropouts.

**Bachelors:** Despite being 24.8% of the total billionaires, they hold only 6% of the total net worth. This indicates that while many billionaires have a bachelor's degree, their average net worth is lower compared to dropouts and those with a master's degree.
